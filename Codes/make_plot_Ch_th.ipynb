# These codes make the theorical variation on hyporheic zone Rn activity as a function of residence time for different initial Rn activities in streamwater

We will start by defining some parameters

```r
th = seq(0.01,100, 0.001) #residence time
ro <- 3000 #Rn production rate
lamda <- 0.18#deacay Constant

#Define the Rn activity in streamwater
c1 <- 5000
c2 <- 2500
c3 <- 1000
c4 <- 500
c5 <- 250
c6 <- 100

#calculate the variations in hyporheic zone Rn as a function of th
Ch1 <- (c1 +ro*th)/(1+lamda*th)
Ch2 <- (c2 +ro*th)/(1+lamda*th)
Ch3 <- (c3 +ro*th)/(1+lamda*th)
Ch4 <- (c4 +ro*th)/(1+lamda*th)
Ch5 <- (c5 +ro*th)/(1+lamda*th)
Ch6 <- (c6 +ro*th)/(1+lamda*th)
```

We can now plot the results

```r

par(mar= c(4,5,1,1), oma = c(0,0,0,0))
plot(Ch1~th, type = "n", log= "xy", xlim = c(0.01,100), ylim = c(100,20000),
     ylab = expression(paste("C"[h]*~"[mBq/L]")), 
     xlab = expression(paste("t"[h]*~"[days]")), 
     xaxt="n", yaxt="n", xaxs = "i", yaxs = "i")
axis(1, at = c(0.01,0.1,1,10,100), labels = paste0(c(0.01,0.1,1,10,100)))
axis(2, at = c(100,1000,10000), labels = paste0(c(100,1000,10000)))

lines(Ch1~th, col = "deepskyblue", lwd = 1.5)
lines(Ch2~th, col = "purple", lty = 1, lwd = 1.5)
lines(Ch3~th, col = "forestgreen", lty = 2, lwd = 1.5)
lines(Ch4~th, col = "blue", lty = 5, lwd = 1.5)
lines(Ch5~th, col = "dodgerblue", lty = 6, lwd = 1.5)
lines(Ch6~th, col = "darkcyan", lty = 1, lwd = 1.5)


legend("bottomright", legend = c("C=5000","C=2500","C=1000", "C=500","C=250","C=100"),
       col = c("deepskyblue","purple","forestgreen","blue","dodgerblue", "darkcyan"),
       lty = c(1,1,2,5,6,1), lwd = 1.5)
```
