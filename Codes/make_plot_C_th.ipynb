# These codes make the theorical relationship between hyporheic zone residence time and Rn activity in a stream with no groundwater inflow

We will start by defining some parameters

```r
ro <- 3000 #Rn production rate
k <- 1 #gas exchange rate
d <- 0.5 #steam depth
phi <- 0.4 #porosity

#h is the hyporheic zone thickness
h1 <- 0.25
h2 <- 0.5
h3 <- 1
h4 <- 0.75

#deacay Constant
lamda <- 0.18

#residence time
th <- seq(0.1,1000, 0.1)

#concentration decay curve if no goundwater is gained
c1 <- (ro*h1*phi)/(((1+lamda*th)*(k+lamda*d))+(lamda*h1*phi))
c2 <- (ro*h2*phi)/(((1+lamda*th)*(k+lamda*d))+(lamda*h2*phi))
c3 <- (ro*h3*phi)/(((1+lamda*th)*(k+lamda*d))+(lamda*h3*phi))
c4 <- (ro*h4*phi)/(((1+lamda*th)*(k+lamda*d))+(lamda*h4*phi))

```

We can now plot the results

```r
par(mar= c(4,5,1,1), oma = c(0,0,0,0))

plot(c3~th, type = "n", log = "xy", xlim = c(0.1,1000), ylim = c(1,1000),
     ylab = expression(paste(""^222*"Rn [mBq/L]")), 
     xlab = expression(paste("t"[h]*~"[days]")), xaxt="n", yaxt="n")


axis(1, at = c(0.1,1,10,100,1000,10000), labels = paste0(c(0.1,1,10,100,1000,10000)))
axis(2, at = c(1,10,100,1000), labels = paste0(c(1,10,100,1000)))

lines(c3~th, xlim = c(0.1,1000),col = "deepskyblue", lwd = 1.5)
lines(c2~th, lty = 2, xlim = c(0.1,1000),col ="blue" , lwd = 1.5)
lines(c1~th, lty = 3, xlim = c(0.1,1000),col = "seagreen4", lwd = 1.5)
lines(c4~th, lty = 4, xlim = c(0.1,1000),col ="slateblue4", lwd = 1.5)
legend("bottomleft", legend = c("h=1", "h=0.75","h=0.5","h=0.25"),lty = c(1,4,2,3),
       col = c("deepskyblue","slateblue4","blue","seagreen4"), lwd = 1.5)
```
