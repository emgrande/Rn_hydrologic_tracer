# These codes make the theorical variation of the ratio of advective transport distance do diffusive transport distance as a function of time for different advective velocities
We will start by defining some parameters

```r

t <- seq(0,1, 0.001) #time in d
Ds <- 6*10^-6 * 86400.00 #effective diffusion coefficient for Rn in wet sediment cm2/d

#advective velocities
v1 <- 1
v2 <- 2
v3 <- 3
v4 <- 4
v5 <- 5
v6 <- 10

La.Ld1 <- v1*sqrt(t/Ds) 
La.Ld2 <- v2*sqrt(t/Ds) 
La.Ld3 <- v3*sqrt(t/Ds) 
La.Ld4 <- v4*sqrt(t/Ds) 
La.Ld5 <- v5*sqrt(t/Ds) 
La.Ld6 <- v6*sqrt(t/Ds) 

```

We can now plot the results

```r

par(mar= c(4,5,1,1), oma = c(0,0,0,0))
plot(t, La.Ld1, type = "l", ylim = c(0,10), col = "deepskyblue", xaxs = "i", yaxs = "i",
     xlab = "Time [d]", 
     ylab = expression(paste("L"[a]*"/L"[d])), lwd = 1.5)
#lines(t, La.Ld2, col = "blue", lwd = 1.5)
lines(t, La.Ld3, col = "darkgreen", lwd = 1.5)
#lines(t, La.Ld4, col = "slateblue4", lty = 5, lwd = 1.5)
lines(t, La.Ld5, col = "purple", lwd = 1.5)
lines(t, La.Ld6, col = "blue", lwd = 1.5)

legend("topright", legend = c("1 cm/d","3 cm/d","5 cm/d","10 cm/d"),
       col = c("deepskyblue","darkgreen","purple","blue"),
       lty = 1, bty = "n",lwd = 1.5)
```
